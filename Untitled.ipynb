```python
import pandas as pd

import seaborn as sns

data=pd.read_csv('Automobile_data.csv')

print(data.head())
```

```python
sns.histplot(data['price'])
```

```python
import matplotlib.pyplot as plt

x=data['bore'].value_counts()

plt.pie(x.values,

       labels=x.index,

        autopct='%1.1f%%')

plt.show()
```

```python
import matplotlib.pyplot as plt

x=data['stroke'].value_counts()

plt.pie(x.values,

       labels=x.index,

        autopct='%1.1f%%')

plt.show()
```

```python
sns.countplot(data['stroke'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

sns.barplot(x=data['stroke'],y=data['horsepower'])

plt.xticks(rotation='horizontal')
```
